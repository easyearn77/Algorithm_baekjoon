```python
#3 힘들었지만 뿌듯★ 

#공백만드는 함수
def m_blank(x,y,num):
  global star #그림 그릴 배열
  len=num//3 #공백 가로&세로 길이
  for i in range(x-len*2+1,x-len+1):
    for j in range(y-len*2+1,y-len+1):
      star[i][j]=' '
  #재귀문(점 갯수가 3개이면 return)
  if num==3:
    return
  else:
    for i in range(x-2*len,x+1,len):
      for j in range(y-2*len,y+1,len):
        m_blank(i,j,len) #재귀문, 1/9영역으로 나눠서 함수 재실행

n=int(input())
star=[['*']*n for _ in range(n)] #별 먼저 찍기

#공백만들기(맨 오른쪽 끝점&점 개수 인수로 전달)
m_blank(n-1,n-1,n)

print('\n'.join([''.join([str(i) for i in row]) for row in star]))
```

```python
#4 (윗 알고리즘 수식 참고)
def move(N,start,to):
  global path
  path.append([start,to])
  global cnt
  cnt+=1

def hanoi(N,start,to,via):
  if N==1:
    move(1,start,to)
  else:
    hanoi(N-1,start,via,to)
    move(N,start,to)
    hanoi(N-1,via,to,start)

path=[]
cnt=0
n=int(input())
hanoi(n,1,3,2)

print(cnt)
for i in path:
  print(i[0],i[1])
```
